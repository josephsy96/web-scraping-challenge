```python
#Import Dependencies
from splinter import Browser
from bs4 import BeautifulSoup as bs
import requests
```

```python
#Cerberus Hemi 
url = "https://astrogeology.usgs.gov/search/map/Mars/Viking/cerberus_enhanced"

c_html = requests.get(url)
```

```python
#Cerberus BeautifulSoup
c_soup = bs(c_html.text,'html.parser')
```

```python
#Pull tile and image
cerb_site = c_soup.find_all('div',{'class':'container'})

cerb_list = []
cerb_image = []

for ct in cerb_site:
    cerb_img_url = ct.find('img',{'class':'wide-image'})['src']
    cerb_title = ct.find('h2',{'class':'title'})
    cerb_list.append(cerb_title.text)
    cerb_image.append(url + cerb_img_url)

print(cerb_list[0])
print(cerb_image[0])
    
```
